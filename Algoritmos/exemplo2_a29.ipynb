```python
# EXEMPLO 2

import pandas as pd
import numpy as np

# obter dados do isp
try:
    # endereço dados do ISP
    ENDERECO_DADOS = "https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv"

    # ler dados do ISP
    df_isp = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')

    # Ver as primeiras linhas (DISPLAY SO EM IPYNB)
    display(df_isp.columns())

except Exception as e:
    print(f"Erro ao obter dados do ISP: {e}")

```

```python
# delimitando as variaveis
try:
    df_roubo_veiculos = df_isp[["cisp","mes","ano","munic","regiao","roubo_veiculo"]]
    # totalizando os roubos de veiculos por municipio, regiao e mes
    

except ImportError as e:
    print(f"Erro ao delimitar variáveis: {e}")
```

```python
# Calculando os Quartis
try:
    array_roubo_veiculos = np.array(df_roubo_veiculos["roubo_veiculo"])

    total = np.sum(array_roubo_veiculos)
    media = np.mean(array_roubo_veiculos)
    mediana = np.median(array_roubo_veiculos)

    # imprimindo os resultados
    print(f"Total de roubos de veículos: {total}")
    print(f"Média de roubos de veículos: {media}")
    print(f"Mediana de roubos de veículos: {mediana}")

    # quartis
    q1 = np.quantile(array_roubo_veiculos, 0.25, method='weibull')
    q3 = np.quantile(array_roubo_veiculos, 0.75, method='weibull')

    print(f"Q1: {q1}")
    print(f"Q3: {q3}")

except Exception as e:
    print(f"Erro ao calcular quartis: {e}")
```

```python
# Criando uma coluna nova, adicionando identificadores
# Concatena a nova coluna no dataframe
try:
    # criando a nova coluna
    # MENOS se for menor que q1, MAIS se for maior que q3 e MAIS OU MENOS se for entre q1 e q3
    df_roubo_veiculos["flag"] = np.where(df_roubo_veiculos["roubo_veiculo"]<=q1,"menos",np.where(df_roubo_veiculos["roubo_veiculo"]>=q3,"mais","mais ou menos"))

    display(df_roubo_veiculos)

    # salvando o dataframe
    df_roubo_veiculos.to_csv("roubo_veiculos_exercicio.csv", sep=';', index=False, encoding='iso-8859-1')

except Exception as e:
    print(f"Erro ao criar coluna nova: {e}")
```
